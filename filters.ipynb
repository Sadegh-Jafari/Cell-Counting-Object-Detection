```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Path to the input image
image_path = "/work/hdd/bczm/sjafarisheshtamad/COMS571/IDCIA v2/images/220815_GFP-AHPC_B_MAP2ab_F1_DAPI_ND1_20x.tiff"  # Replace with your image path

# Load the image
img = cv2.imread(image_path)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  # Convert to RGB for displaying with matplotlib

# Sharpening kernel
sharpening_kernel = np.array([[0, -1, 0],
                               [-1, 5, -1],
                               [0, -1, 0]])

# Apply sharpening filter
sharpened = cv2.filter2D(img, -1, sharpening_kernel)

# Apply deblurring (Gaussian blur)
deblurred = cv2.GaussianBlur(img, (0, 0), 3)

# Combine sharpening and deblurring
combined = cv2.addWeighted(sharpened, 3, deblurred, -1, 0)

# Display the original and filtered images
titles = ['Original Image', 'Sharpened Image', 'Deblurred Image', 'Combined Filters']
images = [img, sharpened, deblurred, combined]

plt.figure(figsize=(12, 8))
for i in range(4):
    plt.subplot(2, 2, i+1)
    plt.imshow(images[i])
    plt.title(titles[i])
    plt.axis('off')
plt.tight_layout()
plt.show()
```
