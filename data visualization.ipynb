```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
 
```

```python
# Paths to the image and ground truth directories
image_dir = '/Users/nazifamouli/Desktop/Research_project/Responsible Ai/IDCIA v2/images'
ground_truth_dir = '/Users/nazifamouli/Desktop/Research_project/Responsible Ai/IDCIA v2/ground_truth'
 
# Get the list of image files
image_files = [f for f in os.listdir(image_dir) if f.endswith('.tiff')]
 
# Initialize a list to store the number of cells per image
num_cells_list = []
 
# Loop through each image and calculate the number of cells
for image_file in image_files:
    # Read the corresponding ground truth CSV
    csv_filename = os.path.splitext(image_file)[0] + '.csv'
    csv_path = os.path.join(ground_truth_dir, csv_filename)
    if not os.path.exists(csv_path):
        print(f"Ground truth CSV not found for {image_file}")
        continue
    ground_truth = pd.read_csv(csv_path)
    # Append the number of cells in this image to the list
    num_cells_list.append(len(ground_truth))
```

```python
# Calculate statistics
num_images = len(num_cells_list)
mean_cells = np.mean(num_cells_list)
median_cells = np.median(num_cells_list)
std_dev_cells = np.std(num_cells_list)

# Display the information
print("Dataset Information:")
print(f"Number of Images: {num_images}")
print(f"Mean Number of Cells per Image: {mean_cells}")
print(f"Median Number of Cells per Image: {median_cells}")
print(f"Standard Deviation of Cells per Image: {std_dev_cells}")
```

```python
sns.set(style='whitegrid')

# Example values for num_images, mean_cells, and std_cells
num_images = len(num_cells_list)
mean_cells = np.mean(num_cells_list)
std_cells = np.std(num_cells_list)

plt.figure(figsize=(8, 6))
plt.hist(num_cells_list, bins=20, color='#1f77b4', edgecolor='black', alpha=0.75)

plt.title('Distribution of Cell Counts per Image for IDCIA v2', fontsize=16, fontweight='bold', pad=15)
plt.xlabel('Number of Cells', fontsize=14, labelpad=10)
plt.ylabel('Frequency', fontsize=14, labelpad=10)
plt.grid(True, linestyle='--', linewidth=0.5, alpha=0.6)

# Display dataset statistics in a text box
plt.text(0.95, 0.95, f'Number of Images: {num_images}\n'
                     f'Mean: {mean_cells:.2f}\n'
                     f'Median: {median_cells:.2f}\n'
                     f'Standard Deviation: {std_cells:.2f}', 
         transform=plt.gca().transAxes, 
         horizontalalignment='right', 
         verticalalignment='top', 
         fontsize=10, 
         bbox=dict(facecolor='white', edgecolor='black', boxstyle='round,pad=0.5'))

plt.tight_layout()
plt.savefig("version2_distribution.png", dpi=300)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Create the boxplot
plt.figure(figsize=(8, 6))  
plt.boxplot(num_cells_list, 
            vert=True,           
            patch_artist=True,    
            boxprops=dict(facecolor='lightblue', color='blue'),
            medianprops=dict(color='red', linewidth=2),
            whiskerprops=dict(color='green', linewidth=2),
            flierprops=dict(marker='o', color='red', markersize=5))

plt.title('Cell Counts per Image IDCIA v2', fontsize=14, fontweight='bold')
plt.ylabel('Number of Cells', fontsize=12)

# Add custom x-ticks
plt.xticks(ticks=[1], labels=['Test Data'], fontsize=11)

plt.grid(True, linestyle='--', alpha=0.7)

# Show the plot
plt.tight_layout()  # Adjust layout to ensure labels fit nicely
plt.savefig("IDCIA_V2.png", format="png", dpi=300)
plt.show()

```

```python
# Calculate statistics
num_images_v1 = len(num_cells_list_v1)
mean_cells_v1 = np.mean(num_cells_list_v1)
median_cells_v1 = np.median(num_cells_list_v1)
std_dev_cells_v1 = np.std(num_cells_list_v1)

# IDCIA V1 Display the information
print("IDCIA V1 Dataset Information:")
print(f"Number of Images: {num_images_v1}")
print(f"Mean Number of Cells per Image: {mean_cells_v1}")
print(f"Median Number of Cells per Image: {median_cells_v1}")
print(f"Standard Deviation of Cells per Image: {std_dev_cells_v1}")
```

```python
train_size = int(0.7 * len(num_cells_list_v1))
val_size = len(num_cells_list_v1)- train_size
test_size = len(num_cells_list)

train_data = num_cells_list_v1[:train_size]
val_data = num_cells_list_v1[train_size:train_size + val_size]
test_data = num_cells_list[:test_size]

# Calculate summary statistics for each dataset
def calculate_statistics(data):
    return {
        'Number of Images': len(data),
        'Mean': np.mean(data),
        'Median': np.median(data),
        'Std Dev': np.std(data)
    }


train_stats = calculate_statistics(train_data)
val_stats = calculate_statistics(val_data)
test_stats = calculate_statistics(test_data)

# Create a table (DataFrame) with the statistics
stats_df = pd.DataFrame({
    'Training': train_stats,
    'Validation': val_stats,
    'Test': test_stats
})

# Transpose to make it easier to read
stats_df = stats_df.T

# Display the statistics in tabular format
display("Summary Statistics:")
display(stats_df)
```
